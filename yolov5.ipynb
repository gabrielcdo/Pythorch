```python
import torch
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
#model = torch.hub.load('ultralytics/yolov5', 'custom', path='best.pt')
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python

color = (255, 0, 0)
cap  = cv2.VideoCapture('/content/drive/MyDrive/13.mp4')

class_names = []
with open("coco.names", "r") as f:
    class_names = [cname.strip() for cname in f.readlines()]
```

```python
while True:
    _,img = cap.read()
    res = model(img)
    rets = []
    for i in res.xyxy[0]:
        print(str(i) + class_names[int(i[5])])
        x1 = int(i[0])
        y1= int(i[1])
        x2 = int(i[2])
        y2 = int(i[3])
        rets.append([[x1,y1,x2,y2], class_names[int(i[5])]])

    for ret,nomes in rets:
        #print(ret[0])
        start = [ret[0],ret[1]]
        end = [ret[2],ret[3]]
        img = cv2.rectangle(img,(start[0], start[1] , end[0], end[1]) ,color,2)
        cv2.putText(img, nomes, (ret[0], ret[1] - 2), 1 , 2 ,(255,255,255), thickness=1,lineType=cv2.LINE_AA)
    #cv2_imshow(img)
    #if cv2.waitKey(1) == 27:
    #    break
cap.release()
cv2.destroyAllWindows()
```
